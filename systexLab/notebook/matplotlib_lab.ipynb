```python
#matplotlib inline
import matplotlib.pyplot as plt
x = range(0,10)
plt.plot(x,x,'r.')
```

```python
import numpy as np
ax = np.array(x)
ay = ax**2
plt.plot(ax, ay, 'g^')
plt.axis([-10,10,-10,100])
```

```python
t = np.arange(0,2,0.1)
plt.plot(t,t,'r+-',t,t**2,'go--',t,t**3,'b*-', linewidth=4)
```

```python
import pandas as pd
randomSequence = pd.DataFrame(np.random.normal(1, 0.08, (100,8)))
accumulates = randomSequence.cumprod()
accumulates.plot()
plt.title('behavior')
plt.xlabel('time')
plt.ylabel('value')
```
